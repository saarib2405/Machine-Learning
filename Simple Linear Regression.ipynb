# Simple Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# Create a Python list containing values
x = [1, 2, 3, 4, 5]
y = [1, 1, 2, 2, 4]
```

```python
#Generate a scatter plot using Matplotlib's scatter function between x and y
plt.scatter(x, y)
plt.xlabel("Advertising Cost")
plt.ylabel("Sales")
plt.show()
```

```python
#Add a constant term to the independent variable x, creating a new array x1.
#This is done because the OLS method typically includes a constant term in the regression equation.
x1 = sm.add_constant(x)

# Fit a linear regression model to the data using OLS
results = sm.OLS(y, x1).fit()

#Print out a summary of the regression results
results.summary()
```

```python
#linear model's predicted values based on given coefficients and input values.
slope = 0.7
intercept = -0.1
def myfunc(x):
  return slope * x + intercept
mymodel = list(map(myfunc, x))
print(slope)
```

```python
print(intercept)
```

```python
print(mymodel)
```

```python
plt.scatter(x,y)
fig = plt.plot(x, mymodel, lw=4, c='red')
plt.xlabel("Advertising Cost")
plt.ylabel("Sales")
plt.show()
```

```python
sales = myfunc(6)
print(sales)
```
